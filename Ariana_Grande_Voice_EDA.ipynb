<b> <h1> Feature Extraction </h1> </b> 

<p> For each audio file in the dataset, we will extract the mel-frequency cepstrum. 

```python
# Getting the features for the audio wav files 
def get_features(wav_name): 
    # 
    if wav_name: 
        X, sample_rate = librosa.load(wav_name, offset=25.0, duration=10)
    #
    loaded_data = librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40)
    scaled_data = np.mean(loaded_data.T, axis=0) 
    # scaled_data = np.array(loaded_data) 
    # Returning the value 
    return scaled_data 

# value = get_features("/home/Blackbox/Documents/MainProjects/Notebooks/ArianaVoiceEDA/ArianaWav/001.wav")
# print(value.shape);
```

```python
# Getting the path to the working directory 
path = os.getcwd() 
# Listing all the folders and files in the working directory 
listedPath = os.listdir(path) 
# Creating a list to hold both the labels and features 
label = list() 
features = list()

# Creating a loop to parse through the files in the woking directory and find specific 
# Directories for further pre-processing. 
ariana_count = 0 
unknown_count = 0 
for labels in listedPath: 
    # If file name folder parsed is "ArianaWav" then set the tag as "Ariana" 
    # And get the full path for the folder 
    if labels == "ArianaWav": 
        tag = "Ariana"
        ariana_realPath = os.path.sep.join([path, labels])
        # After getting the full path to the folder, list all the files and folders in the directory 
        ariana_wav_files = os.listdir(ariana_realPath) 
        # loop through the folder and get each individual wav audio file in a sorted manner. 
        for wav_songs in sorted(ariana_wav_files): 
            # Loading and extracting the features but firstly getting the wav files full path 
            full_wav_path = os.path.sep.join([ariana_realPath, wav_songs])
            # Display each session for the wav file parsed. 
            # print("Extacting Ariana Files: ", wav_songs); 
            # Using a Try and exception method to extract the data features for each parsed wav files and 
            # Return an Exception error if the feature extraction failed. 
            try:
                mfccs = get_features(full_wav_path) 
            # Exception Error 
            except Exception as e: 
                print("Error in Extraction.")
                continue 
            # Appending the value for the tag to the labels list and with its extracted features. 
            label.append(tag); 
            features.append(mfccs);
            # Increasing the count 
            ariana_count += 1 
        
        
    # If file name folder parsed is "OthersWav" then set the tag as "Unknown" and 
    # Get the full path for the folder. 
    elif labels == "OthersWav": 
        tag = "Unknown" 
        unknown_realPath = os.path.sep.join([path, labels])
        # After getting the full path to the folder, list all the files and folders in the directory 
        unknown_wav_files = os.listdir(unknown_realPath) 
        # loop through the folders and get each individual wav audio file in a sorted other. 
        for wav_songs in sorted(unknown_wav_files): 
            # Loading and extracting the features but firstly getting the wav files full path 
            full_wav_path = os.path.sep.join([unknown_realPath, wav_songs])
            # Display each session for the wav file parsed. 
            # print("Extracting Unknown Files: ", wav_songs); 
            # Using the try and exception method to extract the data features for each parsed wav files and 
            # Return an Exception error if the feature extraction failed. 
            try: 
                mfccs = get_features(full_wav_path) 
            # Exception Error 
            except Exception as e: 
                print("Error in Extraction.")
                continue 
            # Appending the value for the tag to the labels list and with its extracted features. 
            label.append(tag); 
            features.append(mfccs);
            # Increasing the count 
            unknown_count += 1 
            
            
    # Moving on...         
    else:
        continue 
        
# Displaying results 
print("Loaded Ariana Audio Of Count: ", ariana_count); 
print("Loaded Others Unknown Of Count: ", unknown_count); 
```

```python
# Displaying the Number of labels and features loaded into memory 
print("Label Length: ", len(label))
print("Features Length: ", len(features))

# Making Input and Output features 
X = np.array(features)
y = np.array(label)

# Scaling down for easy modelling by our neural network 
scaler = StandardScaler() 
scaler.fit_transform(X) 
# Encoding For the input features 
X_scaled = scaler.transform(X) 

# Encoding for the Target OR Label features 
label_scaler = LabelEncoder() 
label_scaler.fit_transform(y) 

# Encoding For the Output features 
y_scaled = label_scaler.transform(y) 
# Converting to Categorical values for easy modelling by our neural network 
y_scaled = to_categorical(y_scaled, 2) 
```

```python
# Saving Encoders for scaler==>Input Features, and 
# label_scaler==> Output Features. 
Encoders = [scaler, label_scaler]

# Setting the name of the encoders pickle serialized file 
filename = 'encoders.pkl' 
pkl.dump(Encoders, open(filename, 'wb'))
```

```python
# Displaying the shape of the encoded values for both input and output features 
print("Scaled Input Shape: ", X_scaled.shape); 
print("Scaled Output Shape: ", y_scaled.shape); 
```

```python
# Splitting the input and output features into train test split 
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, 
                                                   test_size=0.33, 
                                                   random_state=20)

# Displaying the shape of the training and testing data for both the input(X) and output(y) features 
print("Input Shape: {}, {}".format(X_train.shape, X_test.shape)); 
print("Output Shape: {}, {}".format(y_train.shape, y_test.shape)); 
```

```python
# MODEL BUILDING
# Setting the input dimension for the model 
# Input Dimensions for the X_train 
dim = (X_train.shape[1], )

# Output Dimensions for the y_train 
output = y_train.shape[1]

# Initializing the tensorflow sequential model 
model = Sequential() 
model.add(Dense(280, input_shape=dim, activation="relu"))
model.add(Dense(120, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(280, activation="relu"))
model.add(Dense(120, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(output, activation="softmax"))

# Compiling the model 
model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer="adam")

# Model summary 
model.summary() 
```

```python
# Setting the number of epochs 
epoch = 1000 

# Training the model 
H = model.fit(X_train, y_train, batch_size=5, epochs=epoch, validation_data=(X_test, y_test), verbose=0)
```

```python
# Displaying the the training session on a plot 
plt.style.use("ggplot") 

# Getting the accuracy and the validation accuracy 
acc = H.history["accuracy"]
val_acc = H.history["val_accuracy"]

# Getting the loss and the validation loss 
loss = H.history["loss"]
val_loss = H.history["val_loss"] 

# Getting the epochs range 
epoch_range = list(range(len(H.history["accuracy"])))

# Plotting the first graph for the accuracy 
plt.figure(figsize=(20, 8)) 
plt.plot(epoch_range, acc, label = "Training Accuracy") 
plt.plot(epoch_range, val_acc, label = "Validation Accuracy") 
plt.grid(True) 
plt.xlabel("Epochs") 
plt.ylabel("Accuracy") 
plt.legend(loc="lower right") 
plt.title("Training And Validation Accuracy") 
plt.show() 
```

```python
# Plotting the second graph for loss 
plt.figure(figsize=(20, 8))
plt.plot(epoch_range, loss, label = "Training Loss") 
plt.plot(epoch_range, val_loss, label = "Validation Loss") 
plt.xlabel("Epochs") 
plt.ylabel("Loss") 
plt.legend(loc="upper left") 
plt.title("Training And Validation Loss")                        
plt.show(); 
```

```python
# Evaulation of Ariana Model 
results = model.evaluate(X_test, y_test) 
results = results[1] * 100 
# Dispalying the Model Results 
print(results) 
```
